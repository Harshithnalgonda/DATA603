```python
with open('file_1.txt','r',encoding='utf-8') as file:
    text_1 = file.read()
```

```python
with open('file_2.txt','r',encoding='utf-8') as file:
    text_2 = file.read()
```

```python
# some preprocessing for the text
def preprocessing(text):
    """
    This function takes the raw text, cleans the text using regex
    and returns the processed text
    """
    text = contractions.fix(text) # expanding all the contractions in the text like don't - do not 
    text = re.sub(r'[^\w\s]', '', text) # removes all the punctuations and replaces with empty string
    text = text.replace('\n',' ') # replacing new line characters with space
    text = re.sub('\s+',' ',text) #replacing more than one space with single space
    processed_text = text.strip().lower() #removing the leading and trailing spaces from string and making it lowercase
    
    return processed_text
```

```python
processed_text_1 = preprocessing(text_1)
```

```python
processed_text_2 = preprocessing(text_2)
```

#### 1. Write Python code and use MapReduct to count occurrences of each word in the first text file (file.txt). How many times each word is repeated?

```python
class mapreduce_word_counts:
    def __init__(self,text):
        self.text = text
    def mapper(self):
        """
        This mapper function returns the tokenized words list using
        the processed text.
        """
        words = self.text.split(' ')
        return words
    
    def reducer(self):
        """
        This reducer functions returns counts of each word in the form of dictionary in descending order.
        using the tokenized words list.
        """
        words_list = self.mapper()
        counts = dict(Counter(words_list)) #returns the dict with word counts for each word
        sorted_counts = dict(sorted(counts.items(),key=lambda x:x[1],reverse=True)) # sorts the dict by value
        return sorted_counts 
```

```python
mapreduce_1 = mapreduce_word_counts(processed_text_1)
```

```python
mapreduce_1.mapper()
```

```python
mapreduce_1.reducer()
```

#### 2. From the second text file (file2.txt), write Python code and use MapReduct to count how many times non-English words (names, places, spells etc.) were used. List those words and how many times each was repeated.

```python
#https://raw.githubusercontent.com/dwyl/english-words/master/words_alpha.txt
english_words = []
with open('words_alpha.txt','r') as file:
    for word in file:
        english_words.append(word.strip())
```

```python
class mapreduce_non_eng_word_counts:
    def __init__(self,text):
        self.text = text
    def mapper(self,english_words):
        """
        This mapper functions returns the non english words list from the processed text
        using the english words list.
        """
        words = self.text.split(' ')
        non_english_words = []
        for i in words:
            if i not in english_words: #checks if the word is in english words list.
                non_english_words.append(i)
            else:
                continue
        return non_english_words
    
    def reducer(self):
        """
        This reducer functions returns the count of each non english word in descending order.
        The output is in the form of dictionary with key being non english word and value being count of the word.
        """
        words_list = self.mapper(english_words)
        counts = dict(Counter(words_list)) #returns the dict with word counts for each word
        sorted_counts = dict(sorted(counts.items(),key=lambda x:x[1],reverse=True)) #sorts the dict by value
        return sorted_counts
```

```python
mapreduce_2 = mapreduce_non_eng_word_counts(processed_text_2)
```

```python
mapreduce_2.mapper(english_words)
```

```python
mapreduce_2.reducer()
```
